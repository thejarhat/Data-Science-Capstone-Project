```python
import pandas as pd
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',skiprows=0)[0]
df
```

# Data Cleaning 

#### So now we have the data. We will remove the rows with no boroughs assigned and reset the indexes

```python
indexNames = df[ df['Borough'] == 'Not assigned' ].index
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.shape
```

```python
df = df.sort_values('Postal Code', ascending = True)
df
```

```python
coord_df = pd.read_csv("https://cocl.us/Geospatial_data")
coord_df
```

```python
df['Latitude'] = coord_df['Latitude']
df['Longitude'] = coord_df['Longitude']
df.reset_index(drop=True, inplace=True)
df
```
